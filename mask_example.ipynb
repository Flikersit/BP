```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = r'C:\working_space\pilsen_pigs_2023_cvat_backup\workspase\13\data\Tx039D_Vensoubor_00654.jpg'
mask_path = r'C:\working_space\pilsen_pigs_2023_cvat_backup\masks_for_all_organs\13\Tx039D_Vensoubor_00654_mask.png'


image = Image.open(image_path).convert('L')  # Convert to grayscale
mask = Image.open(mask_path).convert('L')   # Convert to grayscale

# Convert images to numpy arrays for processing
image_np = np.array(image)
mask_np = np.array(mask)

# Overlay mask on the original image (e.g., apply the mask on the red channel)
overlay = np.zeros((image_np.shape[0], image_np.shape[1], 3), dtype=np.uint8)
overlay[..., 0] = mask_np  # Red channel for mask
overlay[..., 1] = image_np  # Green channel for the image
overlay[..., 2] = image_np  # Blue channel for the image

# Plot the original image, mask, and overlay
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image_np, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Mask")
plt.imshow(mask_np, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Overlay")
plt.imshow(overlay)
plt.axis('off')

plt.tight_layout()
plt.show()
```
