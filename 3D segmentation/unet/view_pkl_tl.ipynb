```python
import pickle 
import matplotlib.pyplot as plt
import numpy as np
```

```python
path_to_epoch_loss = r'D:\BP\3D segmentation\unet\epoch_loss_values_for_trans_learning.pkl'
path_to_metric = r'D:\BP\3D segmentation\unet\metric_values_for_trans_learning.pkl'

with open(path_to_epoch_loss, 'rb') as loss:
    data_loss = pickle.load(loss)

with open(path_to_metric, 'rb') as metric:
    data_metric = pickle.load(metric)

plt.figure(figsize=(12, 4))

x_loss = np.linspace(1, len(data_loss), len(data_loss))
x_metric = np.linspace(1, len(data_metric), len(data_metric))

plt.subplot(1, 2, 1)
plt.title("Epoch loss")
plt.grid(True)
plt.plot(x_loss, data_loss)

plt.subplot(1, 2, 2)
plt.title("Metric")
plt.grid(True)
plt.plot(x_metric, data_metric)


plt.show
```
